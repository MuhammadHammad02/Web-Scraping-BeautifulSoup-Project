```python
base_url = "https://aheart2help.com/org/california/" # Replace with actual base URL
all_titles = []
page_number = 1

while True:
  url = base_url + "?page=" + str(page_number)  # Adjust URL structure if needed
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')

  title_divs = soup.find_all('div', class_='elementor-element-1f34df1') # Find all title divs
  for title_div in title_divs:
    title_element = title_div.find('h2')
    title = title_element.text.strip() if title_element else None
    all_titles.append(title)

  next_page_button = soup.find('a', class_='next-page') # Adjust class name as needed
  if next_page_button is None:
    break  # Stop scraping if there's no "Next Page" button

  page_number += 1

print(all_titles)
```
